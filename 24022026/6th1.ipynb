```python
class A:
    def __init__(self, a):
        self.a = a
    def __add__(self, ther):
        return self.a + ther.a
    def __sub__(self, other):
        return self.a - other.a
```

```python
class Vechicle:
    def __init__(self, vechile_id, brand):
        self.vechile_id = vechile_id
        self.brand = brand
    def claculate_rent(self):
        print("Price for the vechicle : ")

class Car(Vechicle):
    def __init__(self, vechile_id, brand, price_per_day, day):
        super().__init__(vechile_id, brand)
        self.day = day
        self.price_per_day = price_per_day
    def claculate_rent(self):
        return self.day*self.price_per_day

class Bike(Vechicle):
    def __init__(self, vechile_id, brand, price_per_hour, hour):
        super().__init__(vechile_id, brand)
        self.hour = hour
        self.price_per_hour = price_per_hour
    def claculate_rent(self):
        return self.hour*self.price_per_hour    

class Truck(Vechicle):
    def __init__(self, vechile_id, brand, price_per_km, km):
        super().__init__(vechile_id, brand)
        self.km = km
        self.price_per_km = price_per_km
    def claculate_rent(self):
        return self.km*self.price_per_km
    

vechicle1 = Car("WB123456", "Maruti", 2000, 5)
vechicle2 = Bike("WB123457", "Honda", 1000, 3)
vechicle3 = Truck("WB123458", "TATA", 30000, 10)
print(vechicle1.claculate_rent())
print(vechicle2.claculate_rent())
print(vechicle3.claculate_rent())
```
