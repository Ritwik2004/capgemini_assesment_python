# 18. Which country shows the highest growth in total production over the years? 

```python
gr = df.groupby(['Country','Year'])['Total Production'].sum().reset_index()
```

```python
gr['Growth'] = gr.groupby('Country')['Total Production'].diff()
```

```python
gr.groupby('Country')['Growth'].sum().sort_values(ascending=False).head(1)
```

# 19. Is there any relationship between export price and total production? 

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Export Price'], df['Total Production'])
plt.xlabel("Total Production")
plt.ylabel("Export Price")
plt.title("Relationship between Export Price and Total Production")
plt.show()
```
