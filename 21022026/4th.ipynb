```python
class Bank:
    name = "SBI"
    loc = "Hooghly"
    type = "Government"
    time = "11 - 03"
    def __init__(self, name, age, phone, pan, balance):
        self.name = name
        self.age = age
        self.phone = phone
        self.pan = pan
        self.balance = balance

    def display(self):
        print(self.name, self.loc, self.type)
    def change_phone(self, new_phone):
        self.phone=new_phone
    def withdraw(self,money):
        if money<0:
            print("sorry!!! \nwithdraw money never become negative...")
        else:
            if self.balance<money:
                print("Sorry!!! not enough money...")
            else:
                self.balance -= money
                print("Withdraw successfull! \n your current balance is : ",self.balance)
            
```

```python
c1 = Bank("Ritwik",22,9883398028,"asdf145546",500001)
```

```python
c1.display()
```

```python
c1.withdraw(5000)
```

```python
c1.withdraw(495002)
```

```python
c1.withdraw(-500)
```

```python
class Bank:
    name = "SBI"
    loc = "Hooghly"
    type = "Government"
    time = "11 - 03"
    def __init__(self, name, age, phone, pan, balance):
        self.name = name
        self.age = age
        self.phone = phone
        self.pan = pan
        self.balance = balance

    def display(self):
        print(self.name, self.loc, self.type)
    @classmethod
    def locChanger(self,newloc):
        self.loc = newloc
        print("Updated successfully!!!")
```

```python
c1 = Bank("Ritwik",22,9883398028,"asdf145546",500001)
```

```python
c1.display()
```

```python
c1.locChanger("kolkata")
```
