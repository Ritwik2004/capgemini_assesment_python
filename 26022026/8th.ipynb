```python
# count number of words in a file
with open('data.txt') as f:
    print(len(f.read().split()))
```

```python
# extract the unique word and count them
with open('data.txt') as f:
    print(set(f.read().split()))
    f.seek(0)
    print(len(set(f.read().split())))
```

```python
# create a dictonary of word count
with open('data.txt') as f:
    cnt = {}
    for word in f.read().split():
        if word in cnt:
            cnt[word] += 1
        else:
            cnt[word] = 1
    print(cnt)
```

```python
import csv
with open("data.csv","w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'Course'])
    writer.writerow(['Ritwik', '22', 'IT'])
```

```python
rows = [
    ["Name", "Age"],
    ["Ritwik", 22],
    ["Ritika", 30],
    ["Partha", 22],
]
with open("data.csv","w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)
```

```python
with open("data.csv","w", newline="") as file:
    fieldname = ["name","age"]
    writer = csv.DictWriter(file, fieldnames=fieldname)
    writer.writeheader()
    writer.writerow({"name" : 'Ritwik', "age" : 22})
    writer.writerow({"name" : 'Ritika', "age" : 30})
```

```python
with open("data.csv", "r") as file:
    reader = csv.DictReader(file)
    print(reader) #-> object address
    for row in reader:
        print(row)
```
