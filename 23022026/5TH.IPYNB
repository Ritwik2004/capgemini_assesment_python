```python
# # oops
class Tenth:
    def __init__(self, student, contact, marks):
        self.student = student
        self.contact = contact
        self.marks = marks

    def display(self):
        print(self.student, self.contact, self.marks)
```

```python
class Twelfth(Tenth):   #single inheritance
    def __init__(self, student, contact, marks, twelveMarks, stream):
        super().__init__(student, contact, marks)
        self.twelveMarks = twelveMarks
        self.stream = stream

    def show(self):
        super().display()
        print(self.twelveMarks, self.stream)
```

```python
t1 = Tenth("Ritwik Kundu", 9883398028, "81%")
t1.display()
```

```python
t2 = Twelfth("Ritwik Kundu", 9883398028, "81%", "91%", "Science")
t2.show()
```

```python
class Bachelors(Twelfth):
    def __init__(self, student, contact, marks, twelveMarks, stream, webjeeRank, colgName, department):
        super().__init__(student, contact, marks, twelveMarks, stream)
        self.webjeeRank = webjeeRank
        self.colgName = colgName
        self.department = department
    
    def visit(self):
        super().show()
        print(self.webjeeRank, self.colgName, self.department)
```

```python
t3 = Bachelors("Ritwik Kundu", 9883398028, "81%", "91%", "Science", "20142", "HIT", "IT")
t3.visit()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display_person(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
```

```python
class Student(Person):
    def __init__(self, name, age, isSports, isAcademic):
        super().__init__(name, age)
        self.isSports = isSports
        self.isAcademic = isAcademic
        
    def display_student(self):
        print(f"name: {self.name}")
```

```python
class Teacher(Person):
    def __init__(self, name, age):
        super().__init__(name, age)
    
    def display_teacher(self):
        print(f"Subject: {self.subject}")
```

```python
class Academic(Student):
    def __init__(self, name, age, subject, marks):
        super().__init__(name, age)
        self.subject = subject
        self.marks = marks
        
    def display_academic(self):

        print(f"Subject Marks: {self.marks}")
```

```python
class Sports(Student):
    def __init__(self, name, age, sport_name, level):
        super().__init__(name, age)
        self.sport_name = sport_name
        self.level = level
    
    def display_sports(self):
        print(f"Sport Name: {self.sport_name}")
        print(f"Level: {self.level}")
```

```python
class AllRounderStudent(Academic,Sports):
    def __init__(self, name, age, subject, marks, sport_name, level):
        # Student.__init__(self, name, age)
        Academic.__init__(self, name, age, subject, marks)
        Sports.__init__(self,name, age, sport_name, level)
        
    def display_all(self):
        print("\n--- All Rounder Student Details ---")
        self.display_person()
        self.display_student()
        self.display_academic()
        self.display_sports()
```

```python
obj = AllRounderStudent(
    "Ritwik", 22,"physics",
    95, "Cricket", "college Level"
)

obj.display_all()
```

```python
#######################################################
class Person1:
    def __init__(self, name, age):
        self.name = name
        self.age = age

#hierarchical Inheritance
class Student1(Person1):
    def __init__(self, name, age, rollno):
        super().__init__(name, age)
        self.rollno = rollno
#multiple Inheritance
class Academic1:
    def __init__(self, course = None, cgpa = None):
        self.course = course
        self.cgpa = cgpa

class Sports1:
    def __init__(self,sport_name = None, level = None):
        self.sports_name = sport_name
        self.level = level

#hybrid Inheritance
class AllrounderStudent1(Student1,Academic1, Sports1):
    def __init__(self, name, age, rollno, is_academic=False, is_sports=False, course=None, cgpa=None, sports_name=None, level=None):
        super().__init__(name, age, rollno)
        self.is_academic = is_academic
        self.is_sports = is_sports

        if self.is_academic:
            Academic1.__init__(self,course,cgpa)

        if self.is_sports:
            Sports1.__init__(self, sports_name, level)
            
    def display(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"Roll No : {self.rollno}")
        if self.is_academic:
            print(f"Course : {self.course}")
            print(f"CGPA : {self.cgpa}")
        if self.is_sports:
            print(f"Sport name : {self.sport_nme}")
            print(f"Sport level : {self.level}")

```
